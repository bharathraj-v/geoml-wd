```python
import json

with open('input-google.json') as f:
    data = json.load(f)

data
```

```python
import pandas as pd

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
dates = []

df['clicks'][0][0]['action']['date']

for i in range(len(df)):
    for j in range(len(df['clicks'][i])):
        dates.append(df['clicks'][i][j]['action']['date'])
```

```python
dates[:10]
```

```python
## parse the dates

import datetime

dates = [i[:10] for i in dates]

dates[:10]
```

```python
# parse dates

dates = [datetime.datetime.strptime(i, '%Y-%m-%d') for i in dates]
```

```python
dates[:10]
```

```python
## check the range of years

years = [i.year for i in dates]

years_range = (min(years), max(years))
```

```python
years_range
```

```python
## check the range of months

months = [i.month for i in dates]

months_range = (min(months), max(months))
```

```python
months_range
```
